```python
all_drugs=data_train['drugName'].unique().tolist()
all_drugs=[x.lower() for x in all_drugs]
```

```python
def process_review(review):
    processed_token=[]
    for token in review.split():
        token=''.join(e.lower() for e in token if e.isalnum())
        processed_token.append(token)
    return processed_token
```

```python
input_statement = "What are my options for Naproxen"
```

```python
output = data_train[ data_train['condition']==2]
for x in process_review(input_statement):
    if x in all_drugs:
        print("Found",x)
        conds = data_train[data_train['drugName'].str.lower() == x].condition.unique()
        for cond in conds:
            output.append(data_train[data_train['condition'].str.lower() == cond.lower()])
            print(data_train[data_train['condition'].str.lower() == cond.lower()])
print(output)
```
